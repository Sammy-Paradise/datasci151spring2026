**Q2** For loop with break statement

<font size = "3">

- In the code below, a random array of measurements is generated between 0 and 150.

- Complete the for loop so that when we encounter the first measurement that is strictly larger than the `threshold`, we stop looping over the array.

- After the loop terminates, the variable `first_val_over_threshold` should contain the **value** of the item that exceeded the threshold. (The value, **not** the corresponding index)

````python
measurements = np.random.randint(low = 0, high = 150, size = 300)
threshold = 100
first_val_over_threshold = None


for val in measurements:
    if val > threshold:
        first_val_over_threshold = val
        break


print(first_val_over_threshold)
````

**Q3** For loop with continue statement

<font size = "3">

- Some data is collected, but with some missing values. When there is no data, it is replaced with either the string 
"N/A" or the `None` object.

- Our goal is to calculate the average of the valid data.

- Here is a small example:

````python
data = ['N/A', 4, None, 13, 12]

sum_valid = 4 + 13 + 12
num_valid = 3

print("Average:", sum_valid / num_valid)
````

<font size = "3">

- Complete the code below by adding a for loop with a `continue` statement. At the end of the code, print the average of all valid items in `data`

- **Hint:** You might try writing code in a separate cell that solves the problem for the smaller list in the previous cell. Then take on the larger list for your actual answer.

- **Hint:** Remember `None` is not the same thing as the string `"None"`.

````python
data = [ 12, 7, 3, "N/A", 25, 19, None, 8, 41, 0, 6, 14, 33, "N/A", 9, 22, None, 5, 11, 27]

sum_valid = 0
num_valid = 0


for item in data:
    if item == "N/A" or item is None:
        continue
    sum_valid = sum_valid + item
    num_valid = num_valid + 1


print("The average is:", sum_valid / num_valid)
````

- Suppose we want to see how many experiments it takes to get a **three successes** (in total, not consecutively). Write code that simulates this sequence of experiments. After your code completes, the variable `num_trials` should be equal to the number of experiments it took to achieve the third "Success".

- **Hint:** I've added some code that prevents an infinite loop. Remove it at your own peril...


````python
# finish the code

num_success = 0
num_trials = 0
this_prevents_infinite_loop = 0

while num_success < 3:
    outcome = np.random.choice(["Success", "Failure"], p = [0.25, 0.75])
    num_trials = num_trials + 1
    if outcome == "Success":
        num_success = num_success + 1



    this_prevents_infinite_loop = this_prevents_infinite_loop + 1
    if this_prevents_infinite_loop == 10000000:
        print("The code shouldn't be taking this long. Try again...")
        break



print("Number of trials:", num_trials)
````

**Q5** Flag anomalous sensor readings

<font size = "3">

- In the code cell below, a sequence of numbers representing sensor readings are generated.

- A sensor reading is "out-of-spec" if it is strictly less than 70, or strictly greater than 130.

- Complete the code below, so that the initially empty list `out_of_spec` is filled in with the **Boolean values** `True` and `False`. An entry of `True` means the reading was "out-of-spec"

**Hint:** Sometimes it's easier to figure out how to deal with a smaller problem. You could add a new code cell and set `reading_array` to be:
```python

```
which allows you to quickly verify if your code gives the right output. Then move on to the cell below.

````python
import numpy as np 
reading_array = np.random.normal(loc=100, scale=15, size=60).round(1)

upper_spec = 130
lower_spec = 70
out_of_spec = []


#### complete the code
for reading in reading_array:
    if reading < lower_spec or reading > upper_spec:
        out_of_spec.append(True)
    else:
        out_of_spec.append(False)


print(out_of_spec)


if len(out_of_spec) > 0 and type(out_of_spec) != type(True):
    print("Looks like you didn't fill out_of_spec with Boolean values. Try again")
````

**Q6** Filter out anomalous sensor readings

<font size = "3">

- This is the same idea as Q5, but instead of flagging anomalous readings in a separate list, we overwrite the reading with the Python `None` object.

- So, hypothetically, if item 37 is "out-of-spec" we would want to do
```python
        reading_array[37] = None
```

- Complete the `for` loop below, replacing all "out-of-spec" entries in `reading_array` with the value `None`

**Note:** If you have prior experience with NumPy, it turns out using `np.nan` instead of `None` is equivalent when overwriting entries of a NumPy array.


````python
import numpy as np 
reading_array = np.random.normal(loc=100, scale=15, size=60).round(1)

upper_spec = 130
lower_spec = 70


#### complete the code 
for index in range(len(reading_array)):
    if reading_array[index] < lower_spec or reading_array[index] > upper_spec:
        reading_array[index] = None
````
